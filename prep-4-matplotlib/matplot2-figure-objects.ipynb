```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
fig = plt.figure()
# the default figure is size x*y and has zero axes
# most important parameters are the figsize and dpi

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# small axes
axes2 = fig.add_axes([0.2,0.2,1,0.5])
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# small axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of Four')

# small axes
axes2 = fig.add_axes([1,1,2,2])
# useful for legends outside of the plot

axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,18)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```
