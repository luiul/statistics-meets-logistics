```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2) 
axes[0,0].plot(x,y)
axes[1,1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes: 
    print(ax)
    index = np.where(axes == ax)
    print(index[0][0])
    ax.plot(x,y)
# we can set up the arguments of plot to be list themselves to plot different functions in the graphs
```

```python
# note that there not enough spacing between the axes -> call the .tight_layout() method at the bottom of the cell
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes: 
    ax.plot(x,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

fig.subplots_adjust(wspace=1,hspace=1)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2, figsize=(4,10), dpi=200)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

axes[1,1].set_ylabel('Y Label 1,1')
axes[1,1].set_title('Title 1,1')

fig.suptitle('Figure Level Title', fontsize=16)

plt.tight_layout()

fig.savefig('new_subplots.png', bbox_inches='tight')
```
