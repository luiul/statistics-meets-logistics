```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes: 
    ax.plot(x,y)
# we can set up the arguments of plot to be list themselves to plot different functions in the graphs
```

```python
axes
```

```python
# note that there not enough spacing between the axes -> call the .tight_layout() method at the bottom of the cell
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes: 
    ax.plot(x,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

fig.subplots_adjust(wspace=1,hspace=1)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2, figsize=(4,10), dpi=200)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

axes[1,1].set_ylabel('Y Label 1,1')
axes[1,1].set_title('Title 1,1')

fig.suptitle('Figure Level Title', fontsize=16)

plt.tight_layout()

fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
# LEGEND
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(0.5,0.5))
# for the parameter loc we can pass a string or the numeric codes associated with it; another alternative is to manually pass the tuple of the location (lower left corner)
```

```python
ax.legend(loc=(1.1,0.5))
fig
```

```python
# STYLING
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='lightblue',label='x v x')
# we can also pass an RGB HEX code
ax.plot(x,x**2,color='#FF8C00',label='x v x**2')
ax.plot(x,x**3,color='#8B008B',label='x v x**3')

ax.legend()
```

```python
# STYLING
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='lightblue',label='x v x',lw='10')
# default linewidth is one, the argument is the factor / ratio
ax.plot(x,x**2,color='#FF8C00',label='x v x**2',linewidth='0.25')

ax.legend()
```

```python
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots()

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
```

```python
fig, ax = plt.subplots(figsize=(10,10))

lines = ax.plot(x,x)
lines[0].set_dashes([5,2,10,4])
# input the dashstyle in pairs, the first number is the number of points, the second number is the number of blank points after it (this pattern keeps repeating)
```

```python
type(lines)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

lines = ax.plot(x,x,lw='5')
lines[0].set_dashes([1,2,1,2,10,2])
# input the dashstyle in pairs, the first number is the number of points, the second number is the number of blank points after it (this pattern keeps repeating)
```
