```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)  
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
# plt drew all the points and a connecting line between them
# it also returns the array / list object called  matplotlib.lines.Line2D at a particular memory location
# and rendering the visualiation
```

```python
plt.plot(x,y)
plt.show()
# note that this shows only the visualization without the output above
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('String Title');
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('String Title')
plt.xlim(0,6)
plt.ylim(0,12);
# here we're cutting data points! 
```
