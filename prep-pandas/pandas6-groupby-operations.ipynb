```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./course-data/mpg.csv')
```

```python
df
# mpg = miles per galon 
```

```python
# let's explore the data
# performance change througout the years
# first be decide what's the categorical column here? 
# here: model_year (as we have multiple instances of various cars per year)
# how has performace in mpg change throughout the years -> grab the average mpg per year

# another category can be cylinders (we have multiple instances of 4, 6, 8 cylinder engines)
```

```python
df['model_year'].unique()
```

```python
len(df['model_year'].unique())
# 13 versus 398 rows -> good categorical value 
```

```python
df.groupby('model_year')
# Pandas is waiting for the aggregate call
```

```python
df.groupby('model_year').mean()
# model_year becomes the index of the DataFrame of the aggregate values (columns with numerical values)
```

```python
df.groupby('model_year').mean()['mpg']
```

```python
df.groupby(['model_year','cylinders'])
# we can group by multiple columns
```

```python
df.groupby(['model_year','cylinders']).mean()
# the result is a multi-level / hiearchie index 
# note that model_year and cylinders are NOT column names (second line in the DataFrame)
# to see this call df.groupby(['model_year','cylinders']).mean().columns()

# -> comparison per year * cylinder 
```

```python
df.groupby(['model_year','cylinders']).mean().index
# multi-level index; note that the single elements look like vectors (tuples in Python)
```

```python
df.groupby(['model_year','cylinders']).describe().transpose()
# quick statistical analysis of a DataFrame
```

```python
yc = df.groupby(['model_year', 'cylinders']).mean()
```

```python
yc
```

```python
 yc.index.names
```

```python
yc.index.levels
# outer-levels first
# shows all the possible indices
# note that not all combinations are present in the DataFrame 
```

```python
yc.loc[70]
# we're only interested in 70's
# .loc command work on the most-outside level index and drops the index
```

```python
yc.loc[[70,82]]
```

```python
yc.loc[(70,4)]
# passing a tuple returns the exat row as a Series
```

```python
yc.xs(key=70, level='model_year')
# in only takes a single key as a value!
# to return a DataFrame with multiple keys use yc.loc[[70,82]]
# the cross-section tool is useful for grabbing data based on the inside index
```

```python
yc.xs(key=4, level='cylinders')
# note that the actual level is not shown in the results ->
# assign resulting DataFrame to a descriptive variable, e.g. four_cyl
```

```python
# example: we want the aggregate only for 6 and 8 cylinder cars
# if you're interested in multiple keys, then it's better to filter out irrelevant data
# from the original DataFrame (before calling groupby())

# filter out DataFrame first, and then call groupby() on what we want

six_and_eight_cyl = df[df['cylinders'].isin([6,8])]
six_and_eight_cyl.groupby(['model_year', 'cylinders']).mean()
```

```python
# swap
yc.swaplevel()
```

```python
# sort
yc.sort_index(level='model_year',  ascending=False)
```

```python
# sorting by inside levels may break up the grouping ->
# it usually makes sense to sort by the most outer level
yc.sort_index(level='cylinders',  ascending=False)
```
