```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./course-data/Sales_Funnel_CRM.csv')
# CRM = customer relationship management system
```

```python
df.head()
```

```python
df_obj = df.select_dtypes(['object'])
```

```python
df_obj
```

```python
df[df_obj.columns] = df_obj.apply(lambda obj_data : obj_data.str.strip())
```

```python
df
```

```python
# help(pd.pivot)
```

```python
# Q: how many licenses of each product type did Google purchase? 
# we need the columns: Company, Product and Licenses
lin = df[['Company', 'Product', 'Licenses']]
```

```python
pivot = pd.pivot(data=lin, index='Company', columns='Product', values='Licenses')
```

```python
pivot
```

```python
pivot.loc['Google']
```

```python
# alternative 
df[df['Company']=='Google']
```

```python
# Q: what is the total sum of the licenses and the sum of the sale price of each company ->
# how much have we sold to each company? 
pd.pivot_table(df, index='Company', aggfunc='sum', values=['Licenses','Sale Price'])
```

```python
# alternative
df.groupby('Company').sum().drop('Account Number', axis=1)
```

```python
df
```

```python
# Q: what is the total sum of the sales that an account manager has done per contact
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], aggfunc='sum')
```

```python
# alternative
df.groupby(['Account Manager','Contact']).sum().drop(['Account Number', 'Licenses'], axis=1)
```

```python
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], columns = ['Product'], aggfunc='sum')
```

```python
# we can fill NaN wilh zeros
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], columns = ['Product'], aggfunc='sum', fill_value=0)
```

```python
# alternative
df.groupby(['Account Manager','Contact','Product']).sum().drop(['Account Number', 'Licenses'], axis=1)
```

```python
# we can also pass in multiple direct NumPy calls as the parameter aggfunc
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], columns = ['Product'], aggfunc=[np.sum, np.mean], fill_value=0)
```

```python
# this table might be easier to read
pd.pivot_table(df, index=['Account Manager', 'Contact','Product'], values=['Sale Price'], aggfunc=[np.sum, np.mean], fill_value=0)
```

```python
# we can fill NaN wilh zeros
pd.pivot_table(df, index=['Account Manager', 'Contact','Product'], values=['Sale Price'], aggfunc='sum', fill_value=0, margins=True)
```

```python
df.groupby(['Account Manager','Contact','Product']).describe().drop(['Account Number','Licenses'], axis=1)
```
