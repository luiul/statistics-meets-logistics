```python
df['tip_percentage'] = df['tip'] / df['total_bill'] * 100
df.head()
# assuming the total bill contains the tip
```

```python
df['price_per_person'] = df['total_bill'] / df['size']
df.head()
# this overwrites the values of the existing column
```

```python
df['tip_percentage'] = np.round(df['tip'] / df['total_bill'] * 100,2)
df['price_per_person'] = np.round(df['total_bill'] / df['size'],2)
df
# use universal function available in NumPy to determine how many decimal
# points the data has
```

```python
df.drop('tip_percentage', axis = 1)
# general command to drop row or column
# axis = 0 -> runs through the rows (vertically) for a labeled index to drop
# axis = 1 -> runs through the columns (horizontally) for a labeled variable to drop

# when we call operations / methods on a DataFrame, Pandas does not do them in place, i.e.
# it does not permanently affect the DataFrame 

# to make this change permanent there are two options: 
# 1. set parameter inplace = True in the drop method (will be deprecated)
# 2. assign the ouputs to a DaraFrame object (recommended)
```

```python
df = df.drop('tip_percentage', axis = 1)
df
```
