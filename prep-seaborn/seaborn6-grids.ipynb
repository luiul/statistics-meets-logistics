```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./course-data/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, y='gender',x='math score',kind='box')
```

```python
sns.catplot(data=df, x='gender',y='math score',kind='box',row='lunch')
```

```python
sns.catplot(data=df, x='gender',y='math score',kind='box',row='lunch',col='test preparation course')
```

```python
sns.catplot(data=df, x='writing score',y='math score',kind='violin',row='lunch',col='test preparation course')
```

```python
sns.catplot(data=df, x='writing score',y='math score',row='lunch',col='test preparation course')
```

```python
sns.pairplot(df)
```

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_lower(sns.kdeplot)

g = g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_lower(sns.kdeplot)

g = g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(df,hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_lower(sns.kdeplot)

g = g.map_diag(sns.kdeplot)

g = g.add_legend()
```
