```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./course-data/dm_office_sales.csv')
```

```python
df.head(3)
```

```python
len(df)
```

```python
sns.scatterplot(x='salary',y='sales',data=df)
plt.figure()
```

```python
# seaborn is creating this plot with Matplotlib -> we can edit the figure canvas with plt calls
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education')
# passing a categorical hue automatically generates a legend; in this 2D plot we're showing 3 dimensions of information: (salary, sales, level of education)
```

```python
sns.scatterplot(x='salary',y='sales',data=df, hue='salary')
# passing a continuous hue automatically generates a gradient
```

```python
df.columns
```

```python
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience', palette='viridis')
```

```python
sns.scatterplot(x='salary',y='sales',data=df,size='salary')
# we can change the size of all the points with the parameter s
```

```python
sns.scatterplot(data=df, x='salary', y='sales', s=200, alpha=0.2)
# we can change the transparency of the points with the alpha parameter
```

```python
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)
# linewidth parameter to change the border of the points (Matpotlib)
```

```python
sns.scatterplot(data=df, x='salary', y='sales', style='level of education')
# similar to hue but we diffentiate by style, not by color
```

```python
sns.scatterplot(data=df, x='salary', y='sales', style='level of education',color='red')
```

```python
sns.scatterplot(data=df, x='salary', y='sales', style='level of education',hue='level of education', alpha=0.3, size='salary')
```

```python
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education')
plt.savefig('my_plot.png')
```
