```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./course-data/country_table.csv')
```

```python
df
```

```python
df = df.set_index('Countries')
```

```python
sns.heatmap(df)
# problem: the columns 'life expectancy' is in another unit
```

```python
df = df.drop('Life expectancy', axis=1)
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df,linewidths=0.5)
```

```python
sns.heatmap(df,linewidths=0.5,annot=True)
```

```python
sns.heatmap(df,annot=True,cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.heatmap(df,annot=True,cmap='viridis',linewidth=0.5)
```

```python
plt.figure(dpi=200)
sns.heatmap(df,annot=True,cmap='viridis',linewidth=0.5,center=40)
```

```python
plt.figure(dpi=200)
sns.clustermap(df,annot=True,cmap='viridis',linewidth=0.5)
```

```python
plt.figure(dpi=200)
sns.clustermap(df,annot=True,cmap='viridis',linewidth=0.5,col_cluster=False)
```

```python
plt.figure(dpi=200)
sns.clustermap(df,annot=True,cmap='viridis',linewidth=0.5,col_cluster=False)
# Asia and the World are cluster together <- Asia holds a large majority of the population; Europe and NA are more depeveloped compared to the rest of the regions -> they're cluster together
```
