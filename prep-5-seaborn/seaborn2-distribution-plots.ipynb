```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('./course-data/dm_office_sales.csv')
```

```python
sns.rugplot(data=df,x='salary')
```

```python
sns.displot(data=df, x='salary',bins=100)
```

```python
# stle='dark' to remove the grip
sns.set(style='darkgrid')
sns.displot(data=df, x='salary')
```

```python
# stle='white' to remove the grip
sns.set(style='whitegrid')
sns.displot(data=df, x='salary')
```

```python
# note that we can comment out sns.set() is a high-level command that set the style of the notebook

# sns.set(style='ticks')
sns.displot(data=df, x='salary')
```

```python
# for things like editing color, linewidth, etc. all of the arguments avaiable in Matplotlib customization are available within the displot() call

sns.displot(data=df, x='salary',color='red',edgecolor='blue', linewidth=4,linestyle='--')
```

```python
sns.histplot(data=df, x='salary')
```

```python
sns.displot(data=df, x='salary', kde=True)
```

```python
sns.displot(data=df, x='salary', kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(data=sample_ages, columns=['age'])
```

```python
sample_ages.head(3)
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.set_style('whitegrid')
sns.displot(data=sample_ages, x='age',rug=True)
```

```python
sns.displot(data=sample_ages, x='age',rug=True,bins=30, kde=True)
# note that the KDE here is immediately cut off below 0 and above 100
```

```python
sns.kdeplot(data=sample_ages, x='age')
# note that here (by default) we got some probability outside the existing range
```

```python
sns.kdeplot(data=sample_ages, x='age',clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x='age',clip=[0,100],bw_adjust=0.01)
# small bandwith -> the PDF stack on top of the instances has a very small variance
```

```python
sns.kdeplot(data=sample_ages, x='age',clip=[0,100],bw_adjust=1)
```

```python
sns.kdeplot(data=sample_ages, x='age',clip=[0,100],bw_adjust=1,shade=True)
```
